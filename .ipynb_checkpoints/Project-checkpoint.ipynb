```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.ad.co.il"
headers = {"User-Agent": "Mozilla/5.0"}
```

```python
html = requests.get(base_url)
soup = BeautifulSoup(html.content,'html.parser')
```

```python
tags = soup('div')
for tag in tags:
    print(tag.get_text())
```

```python
field_map = {
    'סוג נכס': ('property_type', str),
    'הנכס': ('property_type', str),
    'שכונה': ('neighborhood', str),
    'כתובת': ('address', str),
    'חדרים': ('room_num', float),
    'מספר חדרים': ('room_num', float),
    'קומה': ('floor', int),
    'שטח בנוי': ('area', int),
    'שטח גינה': ('garden_area', int),
    'ימים לכניסה': ('days_to_enter', int),
    'תשלומים בשנה': ('num_of_payments', int),
    'ארנונה בחודש': ('monthly_arnona', int),
    'ועד בית בחודש': ('building_tax', int),
    'קומות בבניין': ('total_floors', int),
    'תיאור': ('description', str),
    'חניה': ('has_parking', lambda x: 1 if x.strip() in ['יש', 'כן', 'קיימת', 'כן, קיימת'] else 0),
    'מחסן': ('has_stotsge', lambda x: 1 if x.strip() in ['יש', 'כן'] else 0),
    'מעלית': ('elevator', lambda x: 1 if x.strip() in ['יש', 'כן'] else 0),
    'מיזוג': ('ac', lambda x: 1 if x.strip() in ['יש', 'כן'] else 0),
    'נגישות': ('handicap', lambda x: 1 if x.strip() in ['יש', 'כן'] else 0),
    'סורגים': ('has_bars', lambda x: 1 if x.strip() in ['יש', 'כן'] else 0),
    'ממ\'ד': ('has_safe_room', lambda x: 1 if x.strip() in ['יש', 'כן'] else 0),
    'מרפסת': ('has_balcon', lambda x: 1 if x.strip() in ['כן'] else 0),
    'מרוהטת': ('is_furnished', lambda x: 1 if x.strip() in ['כן'] else 0),
    'משופצת': ('is_renovated', lambda x: 1 if x.strip() in ['כן'] else 0),
    'מחיר': ('price', float),
    'מספר תמונות מצורפות': ('num_of_images', int),
    'מרחק ממרכז העיר במטרים': ('distance_from_center', float)
}
```

```python
def extract_listing_urls(page_url):
    resp = requests.get(page_url, headers=headers)
    soup = BeautifulSoup(resp.content, 'html.parser')
    cards = soup.find_all('div', class_='card-block')
    links = []
    for card in cards:
        a_tag = card.find('a', href=True)
        if a_tag and a_tag['href'].startswith('/ad/'):
            links.append(base_url + a_tag['href'])
    return links
```

```python
def extract_apartment_details(ad_url):
    resp = requests.get(ad_url, headers=headers)
    soup = BeautifulSoup(resp.content, 'html.parser')
    records = {}
    for tag in soup.find_all('tr'):
        cells = tag.find_all(['th', 'td'])
        if len(cells) >= 2:
            label = cells[0].get_text(strip=True)
            value = cells[1].get_text(strip=True)
            if label in field_map:
                eng_field, dtype = field_map[label]
                try:
                    val = dtype(value) if not callable(dtype) else dtype(value)
                except:
                    val = None
                records[eng_field] = val
    records['url'] = ad_url
    return records
```

```python
url    = "https://www.ad.co.il/nadlanrent?sp275=17413&sp277=17553,17550"
listing_urls = extract_listing_urls(url)

apartments_data = []
for ad_url in listing_urls:
    print(f"Scraping {ad_url}")
    data = extract_apartment_details(ad_url)
    apartments_data.append(data)
```

```python
url_p2 = "https://www.ad.co.il/nadlanrent?sp275=17413&sp276=17414&sp277=17553,17550&pageindex=2"
listing_urls = extract_listing_urls(url_p2)

for ad_url in listing_urls:
    print(f"Scraping {ad_url}")
    data = extract_apartment_details(ad_url)
    apartments_data.append(data)
```

```python
url_p3 = "https://www.ad.co.il/nadlanrent?sp275=17413&sp276=17414&sp277=17553,17550&pageindex=3"
listing_urls = extract_listing_urls(url_p3)

for ad_url in listing_urls:
    print(f"Scraping {ad_url}")
    data = extract_apartment_details(ad_url)
    apartments_data.append(data)
```

```python
df = pd.DataFrame(apartments_data)
df.to_csv("apartments.csv", index=False)
```

```python
import time
import googlemaps

API_KEY = 'AIzaSyColzcXsNiVmLRjhbjX_XjkiH39JUsnK4c'
DIZENGOFF_SQUARE = "Dizengoff Square, Tel Aviv, Israel"
```

```python
def get_distance_from_center(address, api_key):
    # Build API request
    endpoint = "https://maps.googleapis.com/maps/api/distancematrix/json"
    params = {
        'origins': address + ", תל אביב, ישראל",  # Ensure the address is in TLV
        'destinations': DIZENGOFF_SQUARE,
        'key': api_key,
        'mode': 'walking'  
    }
    response = requests.get(endpoint, params=params)
    data = response.json()
    try:
        distance = data['rows'][0]['elements'][0]['distance']['value']  # in meters
        return distance
    except Exception as e:
        print(f"Error for address {address}: {e}")
        return None
```

```python
# Example: Load your data
df = pd.read_csv("apartments.csv")

# Add a new column for distance
distances = []
for idx, row in df.iterrows():
    address = row['address']
    if pd.isna(address):
        distances.append(None)
        print(f"Skipping row {idx} due to missing address")
        continue
    address_str = str(address)
    dist = get_distance_from_center(address_str, API_KEY)
    print(f"{address_str}: {dist} meters")
    distances.append(dist)
```

```python
df['distance_from_dizengoff'] = distances
df.to_csv("apartments_with_distance.csv", index=False)
print("Done! Distance column added.")
```
